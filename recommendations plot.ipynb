```python
%pylab inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/Cam/Google Drive/Study/2016 research project/plot.csv')

df.index = df.nm_merchant

df = df.drop('nm_merchant', axis=1)
```

```python
# calculate percentages
df1 = pd.DataFrame(100*df.actual / 20000)
df1['series'] = 'actual'
df1.columns = ['Proportion (%)','series']

df2 = pd.DataFrame(100*df['recommended, k=100'] / 20000) #df['recommended, k=100'].sum())
df2['series'] = 'recommended, k=100'
df2.columns = ['Proportion (%)','series']

df3 = pd.DataFrame(100*df['recommended, k=20'] / 20000) #df['recommended, k=20'].sum())
df3['series'] = 'recommended, k=20'
df3.columns = ['Proportion (%)','series']

df4 = pd.concat([df1,df2,df3])
```

```python
df4
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
#f, ax = plt.subplots(figsize=(6, 50))

sns.set_color_codes("pastel")

g = sns.FacetGrid(df4, col="series", size=40, aspect=.1)
g.map(sns.barplot, "Proportion (%)", y=df.index, alpha=0.6);

#plt.yticks(g.get_yticks(), g.get_yticks() * 100)
#plt.ylabel('Proportion [%]', fontsize=16)

#g.set_xticklabels('Proportion [%]', fontsize=16)

titles = ["Entire customer data sample", "Recommendations when k=100", "Recommendations when k=20"]
for ax, title in zip(g.axes.flat, titles):
    ax.set_title(title)

plt.subplots_adjust(top=0.97)
g.fig.suptitle('Merchant penetration in...',fontsize=16)

```
